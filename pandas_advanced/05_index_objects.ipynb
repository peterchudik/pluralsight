# Interval index

```python
# Interval = range
# (0, 1) - open interval
# [0, 1] - closed interval
# [0, 1] - closed on right
```

```python
ii = pd.IntervalIndex.from_breaks([0, 0.5, 1, 1.5, 2, 2.5, 3], closed="right")

ii
```

```python
df_i = pd.DataFrame(data=np.random.randint(low=1, high=7, size=6), index=ii, columns=["data"])

df_i
```

```python
ii = pd.IntervalIndex.from_tuples([(0.0, 0.5), (0.5, 1.0), (1.0, 1.5), (1.5, 2.0), (2.0, 2.5), (2.5, 3.0)], closed="right")

ii
```

```python
df_i = pd.DataFrame(data=np.random.randint(low=1, high=7, size=6), index=ii, columns=["data"])

df_i
```

```python
ii = pd.IntervalIndex.from_arrays([0, 0.5, 1, 1.5, 2, 2.5], [0.5, 1, 1.5, 2, 2.5, 3], closed="right")

ii
```

```python
df_i = pd.DataFrame(data=np.random.randint(low=1, high=7, size=6), index=ii, columns=["data"])

df_i
```

```python
ii
```

```python
ii.contains(0.5)
```

```python
new_interval = pd.Interval(0.5, 1.5, closed="right")

new_interval
```

```python
ii.overlaps(new_interval)
```

```python
df_i[ii.contains(0.5)]
```

```python
df_i[ii.overlaps(new_interval)]
```

# Multilevel index

```python
df_population = pd.DataFrame(data=populations, index=index, columns=["Population"])

df_population
```

```python
# it is not nice to select population for 2010
[i for i in df_population.index if i[1]==2010]
```

```python
index = [("China", 2010), ("China", 2020), ("India", 2010), ("India", 2020), ("USA", 2010), ("USA", 2020)]

populations = [1.34, 1.44, 1.38, 1.18, 0.31, 0.33]
```

```python
multi_index = pd.MultiIndex.from_tuples(tuples=index, names=("country", "year"))

multi_index
```

```python
df_population = pd.DataFrame(data=populations, index=multi_index, columns=["Population"])

df_population
```

```python
df_population.loc["USA", 2010]
```

```python
df_population.loc["China", ]
```

```python
df_population.loc[("China", ) : ]
```

```python
df_population.loc[("China", slice(None)), : ]
```

```python
df_population.loc[(slice(None), 2010), : ]
```

```python
# xs - cross section
df_population.xs("USA", level="country")
```

```python
df_population.xs(2020, level="year")
```
